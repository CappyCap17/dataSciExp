**Experiment 5**

Linechart, Access Label, Legend, Countplot, Line marker

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Salary_Data.csv')
```

```python
df
```

```python
df.head(4)
```

```python
# making a line chart from Years of experience and salary

plt.figure(figsize=(8, 6))
plt.plot(df['YearsExperience'], df['Salary'])
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Line Chart')
plt.show()
```

```python
# making linechart with markers in it

plt.figure(figsize=(8, 6))
plt.plot(df['YearsExperience'], df['Salary'],marker='o')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Line Chart with markers')
plt.show()
```

```python
name_labels = df['YearsExperience']
marks_labels = df['Salary']
plt.figure(figsize=(8, 6))
plt.plot(df['YearsExperience'], df['Salary'])
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Line Chart with access labels')
plt.show()
```

```python
#using countplot for name and age
plt.figure(figsize=(8, 6))
sns.countplot(x='YearsExperience', data=df)
plt.xlabel('Experience')
plt.ylabel('count')
plt.title('Countplot of no. of Years of Experience')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df['YearsExperience'], df['Salary'],marker='o',label="Salary")
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.legend()
plt.title('Line Chart with legend')

plt.legend(title='Legend', loc='upper left')

plt.show()
```
