```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('loan-prediction.csv')
```

```python
data.head(10)
```

```python
color_map = {'Y': 'green', 'N': 'red'}

data['Color'] = data['Loan_Status'].map(color_map)

plt.figure(figsize=(10, 6))
plt.scatter(data['ApplicantIncome'], data['LoanAmount'], c=data['Color'])

plt.xlabel('Applicant Income')
plt.ylabel('Loan Amount')
plt.title('Loan Amount vs Applicant Income by Loan Status')

for status, color in color_map.items():
    plt.scatter([], [], color=color, label=status)

plt.legend(title='Loan Status')
plt.show()
```

```python
grouped_data = data.groupby('Loan_Status').agg({
    'ApplicantIncome': 'mean',
    'LoanAmount': 'mean'
}).reset_index()

bar_width = 0.35
x = range(len(grouped_data))

plt.figure(figsize=(10, 6))
bar1 = plt.bar(x, grouped_data['ApplicantIncome'], width=bar_width, label='Average Applicant Income', color='blue')
bar2 = plt.bar([p + bar_width for p in x], grouped_data['LoanAmount'], width=bar_width, label='Average Loan Amount', color='orange')

plt.xlabel('Loan Status')
plt.ylabel('Average Amount')
plt.title('Average Applicant Income and Loan Amount by Loan Status')
plt.legend()
plt.tight_layout()
plt.show()
```
